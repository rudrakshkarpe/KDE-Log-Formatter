```python
import pandas as pd
```

```python
df = pd.read_csv('Logs/log_sus.csv', sep=';', header=None, names=[ "Start-time","Empty", "Description"])
df.drop(columns=['Empty'], inplace=True)
df.head()
```

```python
df['Start-time'] = pd.to_datetime(df['Start-time'])
df['End-time'] = df['Start-time'].shift(-1)
```

```python
df.head()
```

```python
df['Duration(sec)'] = (df['End-time'] - df['Start-time']).dt.total_seconds()
df['Elapsed(sec)'] = (df['Start-time'] - df['Start-time'].min()).dt.total_seconds()
df.head()
```

```python
df.to_csv('Logs/log_sus_duration.csv', sep=';', header=None, index=False, float_format='%.2f')
```
